### 4.5 Data Mining Methods
A variety of methods are available for performing data mining studies, including **regression**, **classification**, **clustering**, and **association**.

# **Classification**
**Classification:** A key data mining technique that categorizes data based on learned patterns from historical, labeled data.

**Purpose:** Assigns new, unlabeled data to predefined groups or classes.  
**Example:** Predicting weather as “sunny,” “rainy,” or “cloudy.”

**Common Applications:**
- **Credit Approval:** Classifies credit risk (e.g., “good” or “bad”).
- **Store Location:** Rates potential store sites (e.g., “good,” “moderate,” “bad”).
- **Target Marketing:** Identifies likely customers (e.g., “likely” or “no hope”).
- **Fraud Detection:** Flags transactions as fraud (“yes” or “no”).
- **Telecom Churn Prediction:** Predicts if customers will switch providers (“yes” or “no”).

**Types of Prediction Problems:**
- **Classification:** Predicts discrete class labels (e.g., “sunny” or “rainy”).
- **Regression:** Predicts continuous numeric values (e.g., temperature).

**Classification Types:**

- **Binary Classification:** Two possible outcomes (e.g., true/false).  
  **Examples:**
  - **Spam detection:** Is this email spam?
  - **Engagement:** Was the ad clicked?
  - **Document type:** Is this about finance?

- **Multiclass Classification:** More than two possible outcomes.  
  **Examples:** Weather conditions (“sunny,” “rainy,” “cloudy”) or product categories (e.g., “electronics,” “furniture”).


When the output can have multiple values (that is, the predicted label is a categorical variable), the classification is named a **multiclass classification**. Usually, the possible labels are named levels or classes, and the list of them should be finite and known in advance (or else it will be an unsupervised problem, not a supervised one).

Examples of multiclass classification problems are:

- Which kind of flower is this?
- What's the primary topic of this webpage?
- Which kind of network attack am I experiencing?
- Which digit/letter is drawn in the image?

### Classification Model Training

The most common two-step methodology of classification-type prediction involves model development/training and model testing/deployment. In the model development phase, a collection of input data, including the actual class labels, is used. After a model has been trained, the model is tested against the holdout sample for accuracy assessment and eventually deployed for actual use where it is to predict classes of new data instances (where the class label is unknown). Several factors are considered in assessing the model, including the following.

**Predictive accuracy.** The model’s ability to correctly predict the class label of new or previously unseen data. Prediction accuracy is the most commonly used assessment factor for classification models. To compute this measure, actual class labels of a test data set are matched against the class labels predicted by the model. The accuracy can then be computed as the accuracy rate, which is the percentage of test data set samples correctly classified by the model (more on this topic is provided later in the chapter).

**Speed.** The computational costs involved in generating and using the model, where faster is deemed to be better.

**Robustness.** The model’s ability to make reasonably accurate predictions, given noisy data or data with missing and erroneous values.

**Scalability.** The ability to construct a prediction model efficiently given a rather large amount of data.

**Interpretability.** The level of understanding and insight provided by the model (e.g., how and/or what the model concludes on certain predictions).

## Classification Techniques
A number of techniques (or algorithms) are used for classification modeling, including the following:

**Nearest Neighbors.** Use K nearest neighbors to classify observations with unknown labels as a function of their neighbors with similar input features.

**Decision tree analysis**. Decision tree analysis (a machine-learning technique) is arguably the most popular classification technique in the data mining arena.

**Classification via Regression.** Use logistic regression to estimate class probabilities and build classification rules on top of your fitted logistic regression models.

**Multinomial Logistic Regression.** Extend logistic regression to multiclass problems, and use distributed multinomial regression for fast parallel estimation of multinomial logistic regression models.


**Neural networks.** These are among the most popular machine-learning techniques that can be used for classification-type problems.


A complete description of all of these classification techniques is beyond the scope of this book; thus, only several of the most popular ones are presented here.

### K-Nearest Neighbors
The K nearest neighbor (K-NN) algorithm predicts the class for x by asking, “What is the most common class for observations around x?” Figure below shows the K-NN algorithm for a simple illustrative example.

We have a bunch of points in a two-dimensional coordinate system labeled as either a red star or green square.

You want to classify as either red star or green square an unlabeled point marked as “?” that is located near labeled points.

For a specified K, the K-NN algorithm looks at the K nearest labeled points to take a majority vote for what class (label) we should assign to “?”. Figure on the left shows that using K = 3 yields a neighborhood of 2 green squares and 1 red star, so we’d classify the point as a green square. Figure on the right shows that K = 5 yields a neighborhood of 3 red stars and 2 green squares, so we’d classify the point as a red star.

### Calculating Distances
You can think about every observation x as a point in the space of possible values for each variable. The K-NN algorithm counts nearest neighbors based upon the distance from an unlabeled point to other points in this space. The distance K-NN uses is Euclidean distance: the length of a straight line between two points in space.

image.png

This distance is measured in the units of the dimensions of your inputs x, which means that you need to think carefully about what units you are using for these input variables. Unless you have a really good reason to do otherwise, you should convert your inputs to be measured in units of standard deviations, by scaling your data.

```python
# Partition to X and y
X = data.drop(['id', 'class'], axis=1)
```

```python
X.head(3)
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
# Partition to training and testing
from sklearn.model_selection import train_test_split
```

```python
X.shape
```

```python
knn
```

```python
# train the knn model
X_train_dropped = X_train.dropna()
y_train_dropped = y_train[X_train.index.isin(X_train_dropped.index)]


X_test_dropped = X_test.dropna()
y_test_dropped = y_test[X_test.index.isin(X_test_dropped.index)]

# train the knn model with the data after dropping rows
knn.fit(X_train_dropped, y_train_dropped)
```

```python
# test the model
training_score = knn.score(X_train_dropped, y_train_dropped) * 100
training_score
```

```python
test_score = knn.score(X_test_dropped, y_test_dropped) * 100
test_score
```

```python
# Predict the values in X_test
y_pred = knn.predict(X_test_dropped)
y_pred
```

### Classification Decisions

There are two ways to be wrong in a binary problem:
- False Positive (FP): predict ŷ = 1 when y = 0.
- False Negative (FN): predict ŷ = 0 when y = 1.

And there are two ways to be right:
- True Positive (TP): predict ŷ = 1 when y = 1.
- True Negative (TN): predict ŷ = 0 when y = 0.

In general, there will be different `costs` associated with each type of error and different `benefits` associated with each type of success.

A doctor might place a lower cost on overtreatment than on a missed diagnosis, and a higher benefit on catching a true disease case than on correctly detecting an absence of disease.


## Confusion Matrix
Before we start discussing classification diagnostics metrics, we have to introduce an important concept called the **confusion matrix**.

Let's assume that we have two classes and an algorithm that assigns them to an object. Here, the confusion matrix will look like this:

image.png

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test_dropped, y_pred)
print(confusion)
```

```python
TP = confusion[1,1]
TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]

print(TP, TN, FP, FN)
```

Given the importance of understanding the consequences of classification predictions, there is a rich terminology for describing various classification success rates. Below are `common terminology` for different ways to measure the rate of success for a classifier. Recalling that is the predicted class and y is the actual true class:

###  **Precision** or positive predictive value (PPV) is the proportion of true positives among all predicted positives, so cases with ŷ = 1 where true y = 1.

- This metric is defined to be $TP  /  (TP + FP)$.
- It is the proportion of positive results that are true positive results. Positive predictive value
- You can think of precision as **“of those predicted to be positive, how many were actually predicted correctly?”**


```python
# Calculate precision
TP/(TP+FP)
```

```python
from sklearn import metrics
metrics.precision_score(y_test_dropped, y_pred)
```

### **Recall** or sensitivity also known as True Positive Rate (TPR) is the proportion of cases with y = 1 where you have correctly predicted ŷ = 1.


- This metric is defined to be $TP / (TP + FN)$.
- This metric is concerned with the number of correctly predicted positive events.
- You can think of recall as **“of those positive events, how many were predicted correctly?”**

```python
TP/(TP+FN)
```

```python
metrics.recall_score(y_test_dropped, y_pred)
```

### **Specificity** is the proportion of cases with y = 0 where you have correctly predicted ŷ = 0.

```python
TN / (TN + FP)
```

```python
# Classification report
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test_dropped, y_pred, target_names=['malignant', 'benign'])
print(report)
```
