# **Mastering GroupBy in Pandas**

## Introduction to Data Grouping
- Data grouping is a fundamental aspect of data analysis that allows for organizing and summarizing large datasets into meaningful categories.
- This process is essential in extracting valuable insights from data by enabling analysts to perform operations on subsets of data grouped according to certain criteria.

### Definition and Importance of Data Grouping in Analysis
- Data grouping involves categorizing data into segments based on shared attributes, which can then be analyzed independently.
- This approach simplifies complex datasets, making it easier to identify patterns, trends, and anomalies.
- In the context of **Business Intelligence (BI)**, data grouping is instrumental in breaking down business **metrics** by categories such as **time periods**, **geographic regions**, **product lines**, or **customer demographics**, thereby facilitating targeted analysis and strategic decision-making.

### Overview of the GroupBy Operation
- The groupby operation is a powerful tool in data analysis, particularly in the Pandas library for Python.
- It implements the "split-apply-combine" paradigm where data is first split into groups based on one or more keys, an operation is applied to each group independently, and the results are combined back into a structured form.
- This operation is versatile and can be used for a wide range of data manipulation tasks, from simple aggregations to complex transformations and filtrations.


## GroupBy: Split, Apply, Combine

The name `Group by` comes from a command in the SQL database language, but in Pandas you could thing of it in term of  _`split, apply, combine`_. The idea is to apply the same function on subsets of your dataframe, based on some key to split the dataframe in subsets.

The **GroupBy** operation involves the following steps:
- **Splitting** the data into groups based on some criteria
- **Applying** a function to each group independently
- **Combining** the results into a data structure

image.png

````python
# Recreate the DataFrame based on the visual
import pandas as pd
df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'data': [0, 5, 10, 5, 10, 15, 10, 15, 20]
})
````

````python
df
````

**1. Split**

The first step in the process is to "split" the data. This involves dividing the data into groups based on some key(s). In our example, we are splitting the data based on the 'key' column, which contains categorical variables 'A', 'B', and 'C'.

````python
grouped = df.groupby('key')
````

````python
grouped
````

In this code, `groupby('key') `is used to create a GroupBy object. It doesn't actually compute anything until we apply a function. However, it has separated our DataFrame into three groups, one for each unique value in the 'key' column.

**2. Apply**

After splitting the data into groups, the next step is to "apply" a function to each group independently. This could be an aggregation, transformation, or filtration function.

````python
grouped.sum()
````

Here, we apply the sum() function to the 'data' column of each group. The sum() function is one of several built-in aggregate functions that Pandas can apply to groups.

**3. Combine**

The final step is to "combine" the results of the function back into a single DataFrame. When an aggregation function is applied, Pandas does this automatically.

````python
df.groupby('key').sum()
````

This strategy is elegant and powerful, especially when dealing with larger datasets, as it allows for complex operations to be performed on different subsets of data in a concise and readable manner. It's also a testament to the elegant of Pandas, which enables such operations with minimal code.

# Grouping Data in Pandas

## Single Column Grouping

Grouping by a single column is straightforward in Pandas. You select the column you want to group by, and Pandas will create a GroupBy object which can be used to apply aggregate, transformation, or filter operations.

### Syntax and Basic Usage

The syntax for grouping by a single column in Pandas is:

```python
grouped = dataframe.groupby('column_name')


**Load Sales Data**

Let's consider a DataFrame df that records **sales data** for a retail chain. The DataFrame includes the columns: Date, Region, Product_Category, Sales, and Quantity. If we wanted to group this data by Region, we'd use the following code:

````python
sales = pd.read_csv('/content/bi_course_data_aggregation_data.csv')
````

````python
sales.head()
````

**Inspecting GroupBy Objects**

Here are some ways to understand and inspect what's in a GroupBy object:

**Viewing Groups:**

The **.groups** attribute returns a dictionary where each key is a group name (the unique values from the grouping column), and each value is the index labels of the rows belonging to that group.

````python
sales.groupby('Region').groups
````

**Getting a Specific Group:**

The **.get_group()** method allows you to retrieve a group corresponding to a specific key.

````python
sales.groupby('Region').get_group('East')
````

**Number of Groups:**

The **.ngroups** attribute gives you the number of groups.

````python
sales.groupby('Region').ngroups
````

**Looping Over Groups:**

You can iterate over a **GroupBy** object, which returns a tuple with the group name and the data for each group.

````python
for name, group in sales.groupby('Region'):
    print(f"Region: {name}")
    print(group.head(), "\n")
````

**Size of Each Group:**

The **.size()** method returns a Series with the size or count of each group.

````python
sales.groupby('Region').size()
````

**First and Last Entry:**

The **.first()** and **.last()** methods return the first and last row of data for each group, respectively.

````python
sales.groupby('Region').first()
````

````python
sales.groupby('Region').last()
````

**Sampling:**

Sometimes, you may want to take a random sample from each group.

````python
sales.groupby('Region').sample(3) # Gives 3 random samples from every region
````

Understanding and utilizing these pre-aggregation methods allows you to comprehensively explore and manipulate the grouped data.

# **Applying Aggregation Functions**

After you have grouped your data and potentially performed some preliminary operations, the next step in the analysis process is to apply aggregation functions to your groups. Aggregation functions are used to compute summary statistics for each group, which can be especially insightful when analyzing grouped data.

## Common Aggregation Functions

Pandas provides a number of built-in methods for aggregation:

- `.sum()`: Computes the sum of the values for each group.
- `.mean()`: Calculates the mean of the values for each group.
- `.median()`: Finds the median of the values for each group.
- `.min()`: Determines the minimum value within each group.
- `.max()`: Identifies the maximum value within each group.
- `.count()`: Counts the number of non-null values within each group.
- `.std()`: Calculates the standard deviation of the values within each group.

## Custom Aggregations

In addition to the built-in methods, Pandas allows for custom aggregation functions using the `.agg()` method:

- `.agg()`: Can take a function, a string function name, a list of functions, or a dict mapping column names to functions. This flexibility allows you to perform multiple aggregations in a single call and apply specific functions to specific columns.

## Examples of Aggregation

Using the `grouped_by_region` object from the previous section, let's apply some of these aggregation functions.


````python
# Calculate the total sales per region
sales.groupby('Region')['Sales'].sum()
````

````python
# Compute the average quantity sold per region
sales.groupby('Region')['Sales'].mean()
````

````python
# Apply multiple aggregation functions to get summary statistics for sales
sales.groupby('Region')['Sales'].agg(['sum', 'mean', 'std', 'min', 'max'])
````

````python
# Define a custom function to calculate the range of sales
sales.groupby(['Region'])['Sales'].agg(range_sales=lambda x: x.max() - x.min())
````

**Interpreting the Results**

The output from these aggregation functions provides a comprehensive view of the sales data within each region. For instance, the total and average sales can indicate regions with high or low demand, while the range of sales can suggest variability in sales within each region. Standard deviation gives insight into the sales volatility, which can be crucial for inventory and supply chain planning.

By applying these aggregation functions, you can gain a deeper understanding of the regional performance and make data-driven decisions to improve business operations. This kind of analysis is integral to Business Intelligence as it turns raw data into actionable insights.

# Multiple Columns Grouping in Pandas

Grouping by multiple columns in Pandas is an extension of single column grouping. It allows you to explore data with a multi-dimensional approach. When you group by more than one column, you can dissect the dataset across different categorical variables simultaneously. This multi-level grouping can provide a deeper understanding of how combinations of different factors affect your data.

## Grouping by More than One Column

When you group by multiple columns, Pandas creates a hierarchical index (or MultiIndex) for the resulting GroupBy object. This is akin to creating a pivot table where each combination of group keys forms a unique group in the resulting hierarchy.

## Syntax and Basic Usage

The syntax for multiple column grouping is similar to single column grouping but with a list of column names:

```python
grouped = dataframe.groupby(['column_1', 'column_2'])


We want to analyze sales based on both **Region** and **Product_Category**. Here's how you can group by these two dimensions:



````python
# Grouping the DataFrame by 'Region' and 'Product_Category'
sales.groupby(['Region', 'Product_Category']).size()
````

Now, **grouped_by_region_and_category** is a **GroupBy **object that has data grouped by unique combinations of **Region** and **Product_Category**.

````python
# Calculating total sales for each combination of 'Region' and 'Product_Category'
sales.groupby(['Region', 'Product_Category'])['Sales'].sum()
````

````python
# Unstacking the MultiIndex to create a DataFrame
sales.groupby(['Region', 'Product_Category'])['Sales'].sum().unstack(level=['Product_Category'])
````

````python
sales.head()
````

````python
# Creating a pivot table to summarize sales

# Pivot means changing structure of rows to be columns
# Ex: Changing region to be a column and changing produc_category to be a row
sales.pivot_table(values='Quantity', index=['Product_Category'], columns='Region', aggfunc='sum')
````

# **Advanced GroupBy Operations: Window Functions**

Window functions in Pandas are powerful tools for calculating aggregates over rolling or expanding windows. They are particularly useful for analyzing time-series data and deriving insights from sequential observations.

## Understanding Window Functions

Window functions operate on a set of rows and return a single value for each row, preserving the original row structure. They enable complex calculations such as moving averages, cumulative sums, or rank calculations.

## Applying Rolling and Expanding Operations

### **Rolling Operations**

Rolling operations compute statistics (e.g., sum, mean, standard deviation) over a fixed window of observations that moves along with the data. This allows for trend analysis and smoothing techniques.




````python
# Every 3 rows will be averaged
# Creates new column for Rolling_Avg_Sales
sales['Rolling_Avg_Sales'] = sales['Sales'].rolling(window=3, min_periods=1).mean()
````

````python
sales.head()
````

````python
sales['Expanding_Avg_Sales'] = sales['Sales'].expanding().mean()
sales.head()

# Expanding average does not specify a window size because it
# averages as you move ahead.
# While rolling specifies the number of rows to average.
````
