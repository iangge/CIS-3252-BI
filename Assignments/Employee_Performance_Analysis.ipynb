## **Question 1:** Is there a relationship between education level and hours worked?

**Reason for Choosing Box Plot:**  
I chose a Box Plot for this question because education level is nominal and hours worked is numerical. Box plots are effective for measuring numerical differences between groups. In this case, the groups are separated by education level.


```python
# Run Your Statistics
plt.figure(figsize=(10, 6))

# Dictionary for mapping education levels to their string value
education_labels = {
    2: "High School",
    3: "Bachelor's",
    4: "Master's",
    5: "Ph.D./MBA"
}

# Order of education levels from lowest to highest
education_order = ["High School", "Bachelor's", "Master's", "Ph.D./MBA"]

ed_employee_df = employee_df.copy()
ed_employee_df['Education_Level'] = ed_employee_df['Education_Level'].map(education_labels)

sns.boxplot(data=ed_employee_df, x='Education_Level', y='Work_Hours_per_Week', notch=True, order=education_order)

plt.xlabel('Education Level')
plt.ylabel('Work Hours per Week')
plt.title('Relationship between Education Level and Work Hours')
```

```python
# Run Your Visualization
```

#### **Insights:**
Based on the visualization, we can observe that:

*   Ph D and MBA holders work the most hours on average with high school diploma holders having the 2nd highest.
*   High school diploma holders have the largest IQR.
*   Those with advanced degrees tend to have smaller IQRs.

## **Question 2:** Does age affect an employee's salary?

**Reason for Choosing Scatter Plot:**  
I chose a Scatter Plot for this question because age and salary are both numerical values. Scatter Plots allow us to see the type of correlation between two numerical values as well as linear or non-linear trends.


```python
# Run Your Statistics
plt.figure(figsize=(10, 6))

sns.regplot(data=employee_df, x='Age', y='Salary', scatter=True)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Relationship between Age and Salary')
```

```python
# Run Your Visualization
```

#### **Insights:**
Based on the visualization, we can observe that
*    The trend is linear.
*    There is a positive correlation between age and salary.
*    However, this Scatter Plot is overall weak because the dots are widely dispersed from one another.


---

## **Question 3:** How similar are the age ranges in the Human Resources and Engineering departments?

**Reason for Choosing Histograms:**  
I chose Histograms for this question because they are useful for displaying continuous data, such as age. I wanted to compare the distribution of ages between the Human Resources and Engineering departments.


```python
# Run Your Statistics
plt.figure(figsize=(10, 6))

# Creating dataframes from employee ages in Engineering and HR depts
engineering_age = employee_df[employee_df['Department'] == 'Engineering']['Age']
hr_age = employee_df[employee_df['Department'] == 'Human Resources']['Age']

# Creating histograms for the above dataframes
sns.histplot(engineering_age, label='Engineering', kde=True)
sns.histplot(hr_age, label='Human Resources', kde=True)

plt.legend()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Department')
```

```python
# Run Your Visualization
```

#### **Insights:**
Based on the visualization, we can observe that
*    Ages for engineering are slightly right-skewed.
*    Ages for Human Resources are bimodal.
*    Engineering has more employees in the 35-45 age range than Human Resources.
